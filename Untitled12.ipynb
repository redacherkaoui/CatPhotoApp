```python
from datasets import load_dataset

# Load the dataset you were working on
dataset = load_dataset("bigscience/P3", "adversarial_qa_dbert_answer_the_following_q", split="validation[:10%]")
print(dataset[0])
```

```python
def preprocess_function(examples):
    inputs = tokenizer(examples["inputs_pretokenized"], padding="max_length", max_length=128, truncation=True)
    targets = tokenizer(examples["targets_pretokenized"], padding="max_length", max_length=32, truncation=True)
    return {"input_ids": inputs["input_ids"], "attention_mask": inputs["attention_mask"], "labels": targets["input_ids"]}

tokenized_dataset = dataset.map(preprocess_function, batched=True)
print(tokenized_dataset[0])
```

```python
import torch
```

```python
from transformers import AutoModelForSeq2SeqLM

# Initialize the model
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")

# Move the model to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python
from transformers import DataCollatorForSeq2Seq, Seq2SeqTrainingArguments, Seq2SeqTrainer

data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)

training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=3,
    predict_with_generate=True,
    fp16=True if torch.cuda.is_available() else False,
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    eval_dataset=tokenized_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
)
```

```python
# Train the model
trainer.train()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
class BM25:
    def __init__(self, corpus, k1=1.5, b=0.75):
        self.vectorizer = TfidfVectorizer().fit(corpus)
        self.doc_term_matrix = self.vectorizer.transform(corpus).toarray()
        self.k1 = k1
        self.b = b
        self.avg_doc_len = np.mean([np.sum(row) for row in self.doc_term_matrix])

    def compute_bm25(self, query):
        query_vector = self.vectorizer.transform([query]).toarray()[0]
        scores = []
        for doc_vector in self.doc_term_matrix:
            score = 0
            doc_len = np.sum(doc_vector)  # Length of the document
            for term_index, term_weight in enumerate(query_vector):
                if term_weight > 0:
                    f = doc_vector[term_index]  # Term frequency in the document
                    idf = np.log((1 + len(self.doc_term_matrix)) / (1 + np.count_nonzero(self.doc_term_matrix[:, term_index]))) + 1
                    term_score = idf * (f * (self.k1 + 1)) / (f + self.k1 * (1 - self.b + self.b * (doc_len / self.avg_doc_len)))
                    score += term_score
            scores.append(score)
        return np.array(scores)

# Initialize BM25 with the corpus from your dataset
corpus = dataset["inputs_pretokenized"]
bm25 = BM25(corpus)

# Compute BM25 scores for a query
query = dataset[0]["inputs_pretokenized"]
bm25_scores = bm25.compute_bm25(query)
print(bm25_scores)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load a pre-trained sentence embedding model
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Generate embeddings for the query and the documents
query_embedding = embedding_model.encode([dataset[0]["inputs_pretokenized"]])  # Query embedding
doc_embeddings = embedding_model.encode([doc["inputs_pretokenized"] for doc in dataset])  # Embeddings for all documents

# Compute cosine similarity between the query embedding and each document embedding
cosine_similarities = cosine_similarity(query_embedding, doc_embeddings).flatten()

# Sort the documents by their cosine similarity to the query
sorted_doc_indices = np.argsort(-cosine_similarities)  # Negative sign for descending order

# Print top 5 most similar documents
for idx in sorted_doc_indices[:5]:
    python_idx = int(idx)  # Convert numpy.int64 to Python int
    print(f"Document {python_idx} - Cosine Similarity: {cosine_similarities[python_idx]}")
    print(f"Text: {dataset[python_idx]['inputs_pretokenized']}\n")
```

```python
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load the tokenizer and model
tokenizer = T5Tokenizer.from_pretrained("t5-small")
model = T5ForConditionalGeneration.from_pretrained("t5-small")

# Move the model to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# Function to compute perplexity
def compute_perplexity(text):
    inputs = tokenizer(text, return_tensors="pt", max_length=512, truncation=True).to(device)
    with torch.no_grad():
        outputs = model(**inputs, labels=inputs["input_ids"])
    loss = outputs.loss
    perplexity = torch.exp(loss).item()
    return perplexity

# Compute perplexity for each document in the dataset
perplexities = [compute_perplexity(doc["inputs_pretokenized"]) for doc in dataset]

# Sort the documents based on perplexity (lower is better)
sorted_doc_indices = np.argsort(perplexities)

# Print the top 5 most relevant documents based on perplexity
for idx in sorted_doc_indices[:5]:
    print(f"Document {int(idx)} - Perplexity: {perplexities[int(idx)]}")
    print(f"Text: {dataset[int(idx)]['inputs_pretokenized']}\n")
```

```python
# Step 1: Initialize the tokenizer
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("t5-small")

# Step 2: Tokenization Function
def preprocess_function(examples):
    inputs = tokenizer(
        examples["inputs_pretokenized"],
        padding="max_length",
        max_length=128,
        truncation=True
    )
    targets = tokenizer(
        examples["targets_pretokenized"],
        padding="max_length",
        max_length=32,
        truncation=True
    )
    return {
        "input_ids": inputs["input_ids"],
        "attention_mask": inputs["attention_mask"],
        "labels": targets["input_ids"]
    }

# Apply tokenization on the dataset
tokenized_dataset = dataset.map(preprocess_function, batched=True)

# Step 3: Check data structure after tokenization
def check_data_structure(dataset, index=0):
    print(f"Data point {index} structure: {dataset[index]}")

check_data_structure(tokenized_dataset, index=0)

# Step 4: Function to compute gradients and gradient norms
import torch
import numpy as np

# Ensure your model is on the correct device and in training mode
model.to(device)
model.train()  # Enable dropout and other training-specific layers

def compute_gradient_norms(dataset, batch_size=1):
    gradient_norms = []

    for i in range(0, len(dataset), batch_size):
        # Select a batch from the dataset
        batch = dataset.select(range(i, min(i + batch_size, len(dataset))))

        for data_point in batch:
            # Access 'input_ids', 'attention_mask', and 'labels' correctly from the data_point
            input_ids = torch.tensor(data_point['input_ids'], dtype=torch.long).unsqueeze(0).to(device)
            attention_mask = torch.tensor(data_point['attention_mask'], dtype=torch.long).unsqueeze(0).to(device)
            labels = torch.tensor(data_point['labels'], dtype=torch.long).unsqueeze(0).to(device)

            # Zero out gradients
            model.zero_grad()

            # Perform forward pass
            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss

            # Compute gradients with respect to model parameters
            loss.backward()

            # Compute gradient norm
            grad_norm = 0.0
            for param in model.parameters():
                if param.grad is not None:
                    grad_norm += param.grad.data.norm(2).item() ** 2
            grad_norm = grad_norm ** 0.5
            gradient_norms.append(grad_norm)

            # Delete variables to free memory
            del input_ids, attention_mask, labels, outputs, loss
            torch.cuda.empty_cache()

            # Optional: Print progress
            if (i // batch_size) % 100 == 0:
                print(f"Processed {i // batch_size} / {len(dataset) // batch_size} batches")

    return gradient_norms

# Step 5: Compute gradient norms for the entire dataset in batches
# Adjust batch_size as needed (1 is recommended for memory efficiency)
gradient_norms = compute_gradient_norms(tokenized_dataset, batch_size=1)

# Step 6: Select top-N data points based on gradient norms
def select_top_n_data(dataset, gradient_norms, top_n=10):
    # Convert gradient norms to a numpy array for sorting
    gradient_norms = np.array(gradient_norms)

    # Get indices of top-N gradient norms
    top_n_indices = np.argsort(gradient_norms)[-top_n:]

    # Convert numpy.int64 indices to Python int
    top_n_indices = top_n_indices.tolist()
    top_n_indices = [int(idx) for idx in top_n_indices]

    # Retrieve the selected data points
    selected_data = [dataset[idx] for idx in top_n_indices]

    return selected_data


# Example: Select top 10 data points
top_n = 10
selected_data = select_top_n_data(tokenized_dataset, gradient_norms, top_n=top_n)

# Display the selected data
print(f"\nTop {len(selected_data)} data points selected based on gradient norms.")
for i, data_point in enumerate(selected_data, 1):
    print(f"\nData point {i}:")
    print(f"Input (token IDs): {data_point['input_ids']}")
    print(f"Attention Mask: {data_point['attention_mask']}")
    print(f"Labels (token IDs): {data_point['labels']}")
    print(f"Input (pre-tokenized): {data_point['inputs_pretokenized']}")
    print(f"Target (pre-tokenized): {data_point['targets_pretokenized']}")
```
